Single-Layer Perceptron (Linear Data)

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

X, y = make_classification(
    n_samples=200,
    n_features=2,
    n_redundant=0,
    n_clusters_per_class=1,
    class_sep=2,
    random_state=42
)

y = y.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
weights = np.random.randn(2, 1)
bias = 0
lr = 0.01
epochs = 1000

def step(x):
    return np.where(x >= 0, 1, 0)

for _ in range(epochs):
    z = np.dot(X_train, weights) + bias
    y_pred = step(z)

    error = y_train - y_pred
    weights += lr * np.dot(X_train.T, error)
    bias += lr * np.sum(error)

z_test = np.dot(X_test, weights) + bias
y_test_pred = step(z_test)

accuracy = np.mean(y_test_pred == y_test)
print("Linear Dataset Accuracy:", accuracy)
```

Multi-Layer Neural Network (Non-Linear Data)

```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=200, noise=0.2, random_state=42)
y = y.reshape(-1, 1)

input_size = 2
hidden_size = 10
output_size = 1

W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))
W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))

lr = 0.01
epochs = 5000

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

for _ in range(epochs):
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)

    z2 = np.dot(a1, W2) + b2
    y_pred = sigmoid(z2)
    error = y - y_pred
    d_output = error * sigmoid_derivative(y_pred)
    d_hidden = np.dot(d_output, W2.T) * sigmoid_derivative(a1)

    W2 += lr * np.dot(a1.T, d_output)
    b2 += lr * np.sum(d_output, axis=0, keepdims=True)
    W1 += lr * np.dot(X.T, d_hidden)
    b1 += lr * np.sum(d_hidden, axis=0, keepdims=True)

predictions = (y_pred >= 0.5).astype(int)
accuracy = np.mean(predictions == y)
print("Non-Linear Dataset Accuracy:", accuracy)
```
