```python
import numpy as np
import gzip
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.datasets import mnist
import numpy as np

# Load dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize
X_train = X_train.reshape(-1, 784).T / 255.0
X_test = X_test.reshape(-1, 784).T / 255.0
```

```python
input_size = 784
hidden_size = 64
output_size = 10
W1 = np.random.randn(hidden_size, input_size) * 0.01
b1 = np.zeros((hidden_size, 1))
W2 = np.random.randn(output_size, hidden_size) * 0.01
b2 = np.zeros((output_size, 1))
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)

def softmax(z):
    exp = np.exp(z - np.max(z, axis=0))
    return exp / np.sum(exp, axis=0)
```

```python
def forward_pass(X):
    Z1 = W1.dot(X) + b1
    A1 = sigmoid(Z1)
    Z2 = W2.dot(A1) + b2
    A2 = softmax(Z2)
    return Z1, A1, Z2, A2
```

```python
def backprop(X, Y, Z1, A1, Z2, A2):
    m = X.shape[1]

    Y_onehot = np.zeros((10, m))
    Y_onehot[Y, np.arange(m)] = 1

    dZ2 = A2 - Y_onehot
    dW2 = (1/m) * dZ2.dot(A1.T)
    db2 = (1/m) * np.sum(dZ2, axis=1, keepdims=True)

    dA1 = W2.T.dot(dZ2)
    dZ1 = dA1 * sigmoid_derivative(Z1)
    dW1 = (1/m) * dZ1.dot(X.T)
    db1 = (1/m) * np.sum(dZ1, axis=1, keepdims=True)

    return dW1, db1, dW2, db2
```

```python
def update_params(dW1, db1, dW2, db2, lr):
    global W1, b1, W2, b2
    W1 -= lr * dW1
    b1 -= lr * db1
    W2 -= lr * dW2
    b2 -= lr * db2
```

```python
_, _, _, A2_test = forward_pass(X_test)
test_predictions = np.argmax(A2_test, axis=0)
test_accuracy = np.mean(test_predictions == y_test) * 100
print("Test Accuracy:", test_accuracy)
```
